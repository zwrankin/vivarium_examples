## Or, initialize from a configuration and run interactively

```python
from vivarium.interface import setup_simulation_from_model_specification
```

```python
config_file = 'configurations/boids.yaml'
sim = setup_simulation_from_model_specification(config_file)
```

```python
sim.take_steps(5)
```

```python
# Note that in interactive mode, gif doesn't automatically knit, 
# here's how to manually do it 
import imageio
import os
image_directory = 'output/test4' #Needs to match directory of your images, from specification file
images = [i for i in os.listdir(image_directory) if 'png' in i]
images = [f'step_{i}.png' for i in range(1, len(images) + 1)]
with imageio.get_writer(f'{image_directory}/movie.gif', mode='I', fps=2) as writer:
    for i in images:
        image = imageio.imread(f'{image_directory}/{i}', format='png')
        writer.append_data(image)
```

## Or, you can load each component

```python
from vivarium.interface import setup_simulation

from vivarium_examples.boids.population import Population
from vivarium_examples.boids.location import Location
from vivarium_examples.boids.flock import FlockKMeans
from vivarium_examples.boids.infection import Infection
from vivarium_examples.boids.visualization import SaveFrames
```

```python
components = [Population(), Location(), FlockKMeans(), Infection(), SaveFrames('test4', plot_type='cluster')]
sim = setup_simulation(components)
```

```python
sim.take_steps(10)
```
